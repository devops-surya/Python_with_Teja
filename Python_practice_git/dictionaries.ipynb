## Important things to remember in dictionary:
* Keys are unique , cant have same key to two objects 
* Dictionaries can be iterated

````python
# loop
# in the following style we get only key or value once from the dictionary
for company in ceo_dict:
    print(type(company))
    print(company)
````

````python
# loop
# in the following style we get only key or values from the dictionary
for company in ceo_dict:
    print(company, ceo_dict[company])
````

````python
# .items() usage to get both key and value at the same time
for company , ceo in ceo_dict.items():
    print(company, ceo)
````

````python
# keys and values usage 
ceo_dict.keys() # to get all keys from the dictionary
````

````python
ceo_dict.values()
````

````python
list(ceo_dict.keys()) # to get all keys in list format
````

````python
list(ceo_dict.values()) # to get all keys in list format
````

# Difference Between ASCII and Unicode (Beginner Friendly)

## 🧠 Why Unicode came when ASCII already existed

ASCII was created first, but it had a **big limitation**:
- It could only represent **English characters**

As computers spread worldwide, people needed:
- Different languages (Telugu, Hindi, Chinese, etc.)
- Special symbols (₹, €, ©)
- Emojis 😊

👉 ASCII couldn’t handle these.

So **Unicode was created** to support **all languages and symbols globally**.

---

## 🔑 Key Differences

| Feature        | ASCII                          | Unicode                          |
|----------------|--------------------------------|----------------------------------|
| Full Form      | American Standard Code for Information Interchange | Universal Character Encoding Standard |
| Characters     | 128 characters                 | 140,000+ characters              |
| Languages      | English only                  | All languages                    |
| Emoji Support  | ❌ No                          | ✅ Yes                           |
| Storage        | 7 bits                        | UTF-8, UTF-16, UTF-32 (flexible) |

---

## 🔢 Example

| Character | ASCII Value | Unicode Value |
|----------|------------|---------------|
| A        | 65         | U+0041        |
| ₹        | ❌ Not supported | U+20B9 |
| 😊       | ❌ Not supported | U+1F60A |

---

## 💡 Simple Analogy

- ASCII = Small dictionary (only English)
- Unicode = Big global dictionary (all languages + emojis)

---

## 🚀 Final Understanding

- ASCII was **simple and enough** in early days
- But the world needed **more languages and symbols**
- Unicode solved this by creating **one standard for everything**

👉 Today, almost all systems use **Unicode (especially UTF-8)**
