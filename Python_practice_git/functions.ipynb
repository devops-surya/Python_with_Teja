```python
# Define a function which returns a value with docstring and type hints 
def is_even(number:int) -> bool:
    """This function checks the number is even or add 

    Args:
        number (int): number to be verified

    Returns:
        bool: returns true , if even false otherwise
    """
    return number % 2 == 0
```

```python
# Define a function which returns a value with docstring and type hints 
def print_message(message:str, count:int) -> None:
    """This messages prints the message , depends on count

    Args:
        message (str): message to be printed
        count (int): no of times message should be printed
    """
    index = 0 
    while index < count:
        print(message)
        index += 1 

# help(print_message) in RPEL 
```

```python
def add(a: int, b: int) -> int:
    """This function adds two numbers

    Args:
        a (int): first number
        b (int): second number

    Returns:
        int: sum of a and b
    """
    return a + b    
```

```python
# postion argument
add(2, 3)
```

```python
# named argument
add(a=2, b=3)   
```

```python
# optional argument
def power(number, p=2):
    """_summary_

    Args:
        number (_type_): _description_
        exponent (int, optional): _description_. Defaults to 2.

    Returns:
        _type_: _description_
    """
    return number ** p
```

```python
# optional argument
power(3)
```

```python
power(3, 3)
```
