````python
def even_numbers(start:int=0, end:int=10):
    even_numbers_list = []
    while start <= end:
        if start % 2 == 0:
            even_numbers_list.append(start)
        start += 1
    return even_numbers_list
````

````python
def even_numbers_generator(start:int=0, end:int=10):
    while start <= end:
        if start %2 ==0 :
            yield start
        start += 1
````

````python
def is_prime(number:int) -> bool:
    """Checks if the number is prime or not

    Args:
        number (int): number to be checked

    Returns:
        bool: Returns True if prime False otherwise
    """
    is_prime_result = True
    if number < 2:
        is_prime_result = False
    else:
        index = 2
        while index < number:
            if number % index == 0:
                is_prime_result = False
                break
            index += 1
    return is_prime_result
````

````python
prime_numbers = [number for number in range(100, 1000) if is_prime(number)]
print(prime_numbers)
````

````python
numbers = [10,20,30,40,50,60, 7]
# i want to find if there is any prime number in this list
prime_numbers = []
for number in numbers:
    if is_prime(number):
        prime_numbers.append(number)
if len(prime_numbers) > 0:
    print("There are some prime numbers")
else:
    print("There are no prime numbers")
````

````python
# 1. iter() — Convert iterable to iterator:
my_list = [1, 2, 3]
it = iter(my_list)    # converts list to iterator

print(next(it))   # 1
print(next(it))   # 2
print(next(it))   # 3
````

````python
# 2. next() — Get next item from iterator:
it = iter([10, 20, 30])

print(next(it))          # 10
print(next(it))          # 20
print(next(it))          # 30
print(next(it, 'done'))   # "done" ← default value when exhausted
````

````python
# 7. sorted() — Returns sorted list:
print(sorted([3, 1, 2]))              # [1, 2, 3]
print(sorted([3, 1, 2], reverse=True))# [3, 2, 1] ← descending
print(sorted(["banana", "apple"]))    # ['apple', 'banana']

# sort by key
words = ["banana", "apple", "kiwi"]
print(sorted(words, key=len))         # ['kiwi', 'apple', 'banana'] ← by length
````
