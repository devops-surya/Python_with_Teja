## Object oriented programming 
### Class : 
* Class is a blueprint 
### Object : 
* Object is object is an entity which is real (my bank account)
* Flow:
    1. Design a class 
    2. create a object 
    3. Work with object 
* Relationships between Objects
    * On a broader note we have two kinds of relationships
        * Inheritence: is-a relationship
        * Composition: has-a relationship

* Everything is an object in the world . They will have two things :
    1. contents   : what it has 
    2. Characteristics : what it can do 
* Example of understanding bank account as object :
    * Contents : account_number, balance, customer_name
    * characteristics : withdraw, deposit , transfer


* class will have two thing :
    * contents : 
        * [account_number:str, name:str, balance:float] -- These are called members(attributes), which are similar to variables
    * characteristics : 
        * [transfer(amount, account)  , withdraw(amount)] - These are methods(functions)

* Again members will have two levels :
    * class level members
    * instance level menbers (object level)
* Again methods will have two levels:
    * class level methods
    * instance level methods (object level)

### Define class and understand naming convention 

| Element  | Convention | One-line Example            |
| -------- | ---------- | --------------------------- |
| Class    | PascalCase | `class UserAccount: pass`   |
| Variable | snake_case | `user_name = "teja"`        |
| Method   | snake_case | `def get_user(): pass`      |
| Function | snake_case | `def calculate_sum(): pass` |
| Constant | UPPER_CASE | `MAX_LIMIT = 100`           |
| Module   | lowercase  | `file: payment_service.py`  |
| Package  | lowercase  | `folder: services/`         |

## Defining & calling instance level members

````python
## class with class level &  instance level members  :
class NewBankAccount:
    # class level member
    bank_name: str = "HDFC"

# Constructor(initiator for instance level members)
    def __init__(self):
        self.balance= 0
        self.account_number=""
        self.customer_name=""
    def withdraw(amount: float):
        pass
````

````python
# calling instance level member :
user3_account = NewBankAccount()
type(user3_account)
````

````python
print(user3_account.balance)
print(user3_account.account_number)
print(user3_account.customer_name)
````

````python
## Create a bankaccount for user4

class NewBankAccount:
    # class level member
    bank_name: str = "HDFC"

# Constructor(initiator for instance level members)
    def __init__(self, balance:float, account_number:str, customer_name:str):
        self.balance= balance
        self.account_number= account_number
        self.customer_name= customer_name
    def withdraw(self, amount: float) -> float:
        self.balance -= amount
        return self.balance
````

````python
user4 = NewBankAccount(balance=10000, account_number="9865549", customer_name="ramesh")
````

````python
print(user4.balance)
print(user4.account_number)
print(user4.customer_name)
````
