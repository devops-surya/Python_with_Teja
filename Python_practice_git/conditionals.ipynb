```python
# nested if , if in if 
number = 3
if number > 0:
    print("The number is positive")
    if number % 2 == 0:
        print("the number is even")
    else:
        print("The number is odd ")
elif number < 0:
    print("The number is negetive")
else:
    print("The number is zero")
print("Done")
```

```python
salary = 5000000
if salary <= 400000:
    print("No tax . you are in free tax slab")
elif 400001 <= salary <= 800000:
    print(" You need to pay 5% tax")
    tax_payable = (salary * 5)/100
    print(f"You need to pay tax of rupees {tax_payable}")
elif 800001 <= salary <= 1200000:
    print("You need to pay 10% tax")
    tax_payable = (salary * 10)/100
    print(f"You need to pay tax of rupees {tax_payable}")
elif 1200001 <= salary <= 1600000:
    print("you need to pay 15% tax")
    tax_payable = (salary * 15)/100
    print(f"You need to pay tax of rupees {tax_payable}")
elif 1600001 <= salary <= 2000000:
    print("you need to pay 20% tax")
    tax_payable = (salary * 20)/100
    print(f"You need to pay tax of rupees {tax_payable}")
elif 2000001 <= salary <= 2400000:
    print("You need to pay 25* tax")
    tax_payable = (salary * 25)/100
    print(f"You need to pay tax of rupees {tax_payable}")
elif salary > 2400001:
    print("you need to pay 30% tax")
    tax_payable = (salary * 30)/100
    print(f"You need to pay tax of rupees {tax_payable}")

print ("Done")
```
