```python
# collection datat types - list , tuple , set , frozenset, dict 
# list - list of values  , not unique , allow duplicates
x = [1,2,3]
x.append(4)
print(x)
x.remove(4)
print(x)
```

```python
print(x[0])
```

```python
# tuple - it is immutable , cannot change , not unique , allow duplicates
x = (1,2,3)
# x.append(4) - it is immutable , cannot change , append/remove wont work on tupple
print(x)
```

```python
print(x[2])
```

```python
# single value tuple 
x = (5,)
```

```python
# set in python - mutable , can change , but unique and no duplicates
x = {1,2,3}
print(x)
```

```python
x.add(4)
print(x)
x.remove(4)
print(x)
```

```python
# frozenset - immutable , cannot change , but unique and no duplicates
x = frozenset([1,2,3])
print(x)
# append/remove cannot work 
print(type(x))
```

```python
x = (1,2,3)
print(type(x))
x = ([1,2,3])
print(type(x))
x =[1,2,3]
print(type(x))
x = {1,2,3}
print(type(x))

```

```python
# dictionary - dictionaries are one of the most important data types in Python, especially for real-world work . It stores key value pairs 
employee = {
    "name": "Venkata",
    "age": 25,
    "role": "DevOps"
}
print(employee["name"])
```

```python
employee["role"] = "Devops/ AI "
print(employee["role"])
```

```python
employee["experience"] = 10 
print(employee["experience"])
print(employee)
```

| Feature / Type | List | Tuple | Set | Frozenset | Dictionary |
|---------------|------|-------|-----|-----------|------------|
| Syntax | [] | () | {} | frozenset() | {key: value} |
| Ordered | ✅ Yes | ✅ Yes | ❌ No | ❌ No | ✅ Yes (Py3.7+) |
| Mutable | ✅ Yes | ❌ No | ✅ Yes | ❌ No | ✅ Yes |
| Allows Duplicates | ✅ Yes | ✅ Yes | ❌ No | ❌ No | ❌ Keys: No / Values: Yes |
| Indexing | ✅ Yes | ✅ Yes | ❌ No | ❌ No | ❌ (use keys) |
| Key-Value Pair | ❌ No | ❌ No | ❌ No | ❌ No | ✅ Yes |
| Can be Dictionary Key | ❌ No | ✅ Yes | ❌ No | ✅ Yes | ❌ No |
| Add Elements | append(), extend() | ❌ Not allowed | add(), update() | ❌ Not allowed | key = value |
| Remove Elements | remove(), pop() | ❌ Not allowed | remove(), discard() | ❌ Not allowed | pop(), del |
| Performance | Medium | Fast | Fast (lookup) | Fast (read-only) | Very fast (lookup) |


| Type | Best Use Case |
|------|--------------|
| List | When you need ordered, changeable data (e.g., user inputs, logs) |
| Tuple | Fixed data that should not change (e.g., coordinates, function returns) |
| Set | Remove duplicates, fast membership checks |
| Frozenset | Immutable set, safe data, dictionary keys |
| Dictionary | Structured data (APIs, configs, agent inputs/outputs) |
